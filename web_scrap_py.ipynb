```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.text)
```

##Table One


```python
#soup.find('table', class_ = 'wikitable sortable')
table = soup.find_all("table")[1]
```

```python
titles = table.find_all("th")
```

```python
titles = [title.text.strip() for title in titles]
print(titles)
```

```python
df = pd.DataFrame(columns = titles)
df
```

```python
column_data=table.find_all("tr")
```

```python
for column in column_data[1:]:
    col = [row.text.strip() for row in column]
    col = [data for data in col if data != '' ]

    length = len(df)
    df.loc[length] = col
```

```python
df
```

```python
table_two=soup.find_all('table')[2]
```

```python
table_two.find_all('th')
```

```python
t2_titles = [title.text.strip() for title in table_two.find_all('th')]
```

```python
table2_rows = table_two.find_all('tr')[1:]
```

```python
table_two = pd.DataFrame(columns = t2_titles)
table_two
```

```python
for row in table2_rows:
    col = [data.text.strip() for data in row]
    col = [data for data in col if data != '']

    l = len(table_two)
    table_two.loc[l] = col
```
